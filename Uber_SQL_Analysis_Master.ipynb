```
def run_query(sql, title="Results"):
    import pandas as pd

    result = pd.read_sql_query(sql, conn)

    print("\n" + "="*50)
    print(title)
    print("="*50)

    print(result.to_string(index=False))

    return result
```

```
run_query("""
SELECT
    Status,
    COUNT(*) AS Total_Trips
FROM trips
GROUP BY Status
ORDER BY Total_Trips DESC
""", title="Query 1 - Trip Count by Status")
```

```
run_query("""
SELECT
    Request_Hour,
    COUNT(*) AS Unfulfilled_Requests
FROM trips
WHERE Status IN ('Cancelled', 'No Cars Available')
GROUP BY Request_Hour
ORDER BY Request_Hour
""", title="Query 2 - Unfulfilled Requests Per Hour")
```

```
run_query("""
SELECT
    Time_Slot,
    COUNT(*) AS Unfulfilled_Requests
FROM trips
WHERE Status IN ('Cancelled', 'No Cars Available')
GROUP BY Time_Slot
ORDER BY Unfulfilled_Requests DESC
""", title="Query 3 - Supply Demand Gap by Time of Day")
```

```
run_query("""
SELECT
    Pickup_Point_Proper,
    Status,
    COUNT(*) AS Total_Requests
FROM trips
WHERE Status IN ('Cancelled', 'No Cars Available')
GROUP BY Pickup_Point_Proper, Status
ORDER BY Pickup_Point_Proper, Total_Requests DESC
""", title="Query 4 - Airport vs City Cancelled and No Cars")
```

```
run_query("""
SELECT
    Driver_id,
    COUNT(*) AS Cancellations
FROM trips
WHERE Status = 'Cancelled'
    AND Driver_id IS NOT NULL
    AND Driver_id <> ''
GROUP BY Driver_id
ORDER BY Cancellations DESC
LIMIT 10
""", title="Query 5 - Top 10 Drivers by Cancellations")
```

```
run_query("""
SELECT
    Pickup_point,
    COUNT(*) AS Total_Requests,
    SUM(CASE WHEN Status = 'Trip Completed' THEN 1 ELSE 0 END) AS Completed,
    ROUND(
        100.0 * SUM(CASE WHEN Status = 'Trip Completed' THEN 1 ELSE 0 END)
        / COUNT(*),
        1
    ) AS Completion_Rate_Pct
FROM trips
GROUP BY Pickup_point
ORDER BY Completion_Rate_Pct DESC
""", title="Query 6 - Completion Rate Airport vs City")
```

```
run_query("""
SELECT
    Pickup_point,
    Time_Slot,
    COUNT(*) AS Problem_Requests
FROM trips
WHERE Status IN ('Cancelled', 'No Cars Available')
GROUP BY Pickup_point, Time_Slot
ORDER BY Problem_Requests DESC
LIMIT 10
""", title="Query 7 - Worst Pickup Point and Time Slot Combos")
```
